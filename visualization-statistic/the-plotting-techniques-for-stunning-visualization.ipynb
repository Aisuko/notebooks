# Creating Visually Stunning Plots

When it comes to crafting plots that mesmerize and leave a lasting impression, customization is key.

## Unleashing the Power of Color Palettes

Color palettes wield incredible power in enhancing the overall visual impact of our plots. Here are some tips for selecting harmonious and visually appealing color schemes:

* Color Schemes
* Contrasting Brilliance
* Color Harmony
* Accessibility

```python
import seaborn as sns
import matplotlib.pyplot as plt

# setting a color palette
sns.set_palette('husl')

# sample data
data=sns.load_dataset('iris')

# Scatter plot with custome color palette
sns.scatterplot(
    data=data,
    x='sepal_length',
    y='sepal_width', 
    hue='species', 
    palette='husl')

plt.title('Scatter Plot with Custom Color palette')
plt.show()
```

## Infusing Character with Line Styles and Markers

* Line Styles
* Markers
* Mix and Match

```python
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, 1, 8, 27, 64, 125]

# Line styles and markers
plt.plot(x, y1, linestyle='--', marker='o', label='Squared')
plt.plot(x, y2, linestyle='-', marker='x', label='Cubed')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Styles and Markers Example')
plt.legend()
plt.show()
```

# Scatter Plots with a Twist: Dynamic Size and Color

Scatter plots have always been a reliable choice for showcasing relatinships between two variables.

```python
import plotly.express as px

# Sample data
df = px.data.iris()

# Scatter plot with dynamic size and color
fig = px.scatter(df, x="sepal_width", y="sepal_length", size="petal_length", color="species",
                 title="Scatter Plot with Dynamic Size and Color")
fig.show()
```

# Animated Visualizations

```python
import plotly.graph_objs as go
import numpy as np

# Sample data
t = np.linspace(0, 10, 50)
x = np.sin(t)
y = np.cos(t)
z = t

# Creating the animated scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers')])

fig.update_layout(scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Z'),
    title='3D Animated Scatter Plot')

fig.show()
```

# Faceing the Small Multiples: Comparison

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset("tips")

# Create a FacetGrid
g = sns.FacetGrid(data, col="day", hue="sex")
g.map(sns.scatterplot, "total_bill", "tip")

# Add legend and titles
g.add_legend()
g.set_titles("{col_name} Day")

plt.show()
```

# Parallel Coordinates Plot

```python
import plotly.express as px

# Load sample data
df = px.data.iris()

# Create a parallel coordinates plot
fig = px.parallel_coordinates(df, color="species_id",
                              dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"],
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)

fig.update_layout(title='Parallel Coordinates Plot for Iris Dataset')
fig.show()
```
