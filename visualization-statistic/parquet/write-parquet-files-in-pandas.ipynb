# Overview

**The Apache Parquest format is a column-oriented data file format.** This means data are stored based on columns, rather than by rows. The benefits of this include significantly faster access to data, especially when querying only a subset of columns. This is because only particular can be read, rather than entire records.

```python
import pandas as pd

df=pd.read_csv("/kaggle/input/b-cell-receptors/b_cell_receptors.csv", header=1)
df.info()
```

```python
df.head(5)
```

# Filter

```python
bcr = df.loc[df["Organism IRI"] == "http://purl.obolibrary.org/obo/NCBITaxon_9606"]
bcr=bcr.dropna(subset="Source Organism")
bcr["Source Organism"].value_counts()[:20]
```

## COVID BCRs

```python
covid = bcr.loc[bcr["Source Organism"] == "SARS-CoV2"]
covid["target"] = "SARS-CoV2"
covid.shape
```

## HIV-1 BCRs

```python
hiv1 = bcr.loc[bcr["Source Organism"].str.contains("HIV-1")]
hiv1["target"] = "HIV-1"
hiv1.shape
```

## Concat both

```python
data=pd.concat([covid, hiv1])
data.shape
```

```python
# use heavy chain sequence only and remove duplicated sequences
data["sequence"]=data["Protein Sequence"]
data=data.dropna(subset="sequence").drop_duplicates("sequence")
data["Source Organism"].value_counts()
```

## Subset Columns

```python
data=data.loc[:, ["sequence", "target"]]
data.head()
```

```python
data["target"].value_counts()
```

# Make the dataset balanced

```python
data=pd.concat([
    data.loc[data["target"]=="SARS-CoV2"].sample(n=228, random_state=0),
    data.loc[data["target"]=="HIV-1"],
])

data["target"].value_counts()
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(6,4), dpi=300)
ax=sns.countplot(x="target", data=data, hue="target")
plt.title("Target distribution")
plt.ylabel("# of Sequences")
sns.despine()
```

# Save data to Apache Parquest Format

```python
from sklearn.model_selection import train_test_split

train, test=train_test_split(
    data, stratify=data["target"], random_state=0, test_size=0.2
)

train.shape, test.shape
```

```python
data.to_parquet("bcr.parquet")
train.to_parquet("bcr_train.parquet")
test.to_parquet("bcr_test.parquet")
```
