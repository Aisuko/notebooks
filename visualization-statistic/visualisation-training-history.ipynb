# Overview

Let's visualize the training history of our model in [Building Transformer Model For Proteins](https://www.kaggle.com/code/aisuko/building-transformer-model-for-proteins)

```python
import pandas as pd

df=pd.read_csv("/kaggle/input/building-transformer-model-for-proteins/results.csv")
df.info()
```

```python
df.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(6,4), dpi=300)
sns.lineplot(data=df, x="Epoch", y="train_loss", label="Train Loss")
sns.lineplot(data=df, x="Epoch", y="val_loss", label="Valid Loss")
plt.ylabel("Cross-Entropy Loss")
```
