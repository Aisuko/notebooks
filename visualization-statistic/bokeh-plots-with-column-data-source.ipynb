# Overview

I'd like to use bokeh do visualization and show it in Vue.js. And I prefer to choice showing the data without Bokeh Server. This is an example, and do remmeber this is only for numberical data. We used many models(OneHotEncoder) from sklearn to convert object to numberical values in [Clustering AI Topcis](https://www.kaggle.com/code/aisuko/clustering-ai-topics) and [visualization-with-pca-and-t-sne](https://www.kaggle.com/code/aisuko/visualization-with-pca-and-t-sne) However, we did not measure the performance and accuracy of it.

```python
import os
import pandas as pd

# Note: see https://github.com/bokeh/bkcharts, this project is no longer maintained
# from bokeh.charts import Bar, output_notebook, show

os.environ['DATASET']='/kaggle/input/diamonds/diamonds.csv'


df=pd.read_csv(os.getenv('DATASET')).drop("Unnamed: 0", axis=1)
df.info()
```

```python
df=df.sample(300, random_state=420)
df.head()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# For invoke bokeh.js in notebook
output_notebook()

p=figure(width=800, height=400)

# because the selected columns are numberical, so we can so this.
p.circle(df.carat, df.price, size=20, color='navy', alpha=0.2)
show(p)
```

```python
import numpy as np

p=figure(width=800, height=400, title="Distribution of Price")


hist, edges=np.histogram(df.price, density=True, bins=50)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color="white")

show(p)
```
