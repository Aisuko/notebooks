# Overview
Here we do data summarization for icu output events features.

```python
import pandas as pd
```

```python
out = pd.read_csv("/kaggle/input/mimic-iv-v2-0-icu-output-events-data/preproc_out_icu.csv",header=0)
freq=out.groupby(['stay_id','itemid']).size().reset_index(name="mean_frequency")
freq=freq.groupby(['itemid'])['mean_frequency'].mean().reset_index()
total=out.groupby('itemid').size().reset_index(name="total_count")
summary=pd.merge(freq,total,on='itemid',how='right')
summary=summary.fillna(0)
#summary.to_csv('out_summary.csv',index=False)
summary['itemid'].to_csv('out_features.csv',index=False)
```
