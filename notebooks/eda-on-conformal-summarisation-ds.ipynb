We will use CHG to score each input_sentence then select sentences via threshold(CIS) and compare to suummary_sentences using **ROUGE/coverage**. This is which we keeps our experiment **focused on causal importance**.

```python
# Keep only the useful columns for CHG + CIS summarization experiment
useful_cols = [
    "input",
    "summary",
    "input_sentences",
    "summary_sentences"
]

df = df[useful_cols]
df.head()
```

```python
from rich.console import Console
from rich.text import Text
import re

def highlight_summary_sentences(paragraph, summary_sentences):
    console = Console()
    text = Text(paragraph)

    for sentence in summary_sentences:
        sentence = sentence.strip()
        if not sentence:
            continue
        # Use regex to find all non-overlapping exact matches
        for match in re.finditer(re.escape(sentence), paragraph):
            start, end = match.span()
            text.stylize("bold red", start, end)

    console.print(text)

# Example usage:
row = df.iloc[0]
paragraph = row["input"]
summary_sentences = row["summary_sentences"]  # Already a list in your data

highlight_summary_sentences(paragraph, summary_sentences)
```
