```python
import torch
from transformers import pipeline

# we use default model is distilbert
pipe=pipeline(task='question-answering',
              model="distilbert/distilbert-base-cased-distilled-squad", 
              device='cuda', 
              torch_dtype=torch.bfloat16)
pipe.device
```

```python
context='Melbourne is the one of the most lievable cities in Earth, however, many incidents there days'
question='Does Melboune is a liveable city?'

# https://github.com/huggingface/transformers/blob/v4.40.0/src/transformers/pipelines/question_answering.py#L224
pipe(context=context, question=question)
```
